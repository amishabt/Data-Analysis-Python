```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("2018 Data.xlsx")
```

```python
df.columns
```

```python
plt.scatter(df['Mean Family Income'], df['Perc with health Insurance'], color='green', marker='*')

# Adding labels and title
plt.xlabel('Mean Family Income')
plt.ylabel('Percentage with Health Insurance')
plt.title('Mean Family Income vs. Percentage with Health Insurance (2018)')
plt.show()
```
