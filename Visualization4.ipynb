```python
import pandas as pd
```

```python
df=pd.read_excel("Simplified Car Sales Data.xlsx")
```

```python
df.head()
```

```python
filter_countries = ['Germany', 'Spain', 'Switzerland']
filter_countries_df= df[df['CountryName'].isin(filter_countries)]
filter_countries_df
```

```python
country_totals = filter_countries_df.groupby('CountryName').sum()
country_totals
```

```python
country_totals=country_totals.drop('LaborCost',axis=1)
country_totals.columns=['Total Sales Price','Total Cost Price']
```

```python
country_totals.plot(kind='barh', color=['purple', 'orange'])
```
