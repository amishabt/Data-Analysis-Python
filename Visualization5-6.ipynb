```python
import pandas as pd

df_2020 = pd.read_csv('2020 Cases Only.csv')
```

```python
df_2020 = pd.read_csv('2020 Cases Only.csv', low_memory=False)
df_2021 = pd.read_csv('2021 Cases Only.csv', low_memory = False)
```

```python
df_all = pd.concat([df_2020, df_2021])
```

```python
df_all.head()
```

```python
from datetime import date
df_all['EventDate']=pd.to_datetime(df_all.EventDate, format="%m/%d/%Y")
```

```python
df_all.head()
```

```python
df_all.dtypes
```

```python
# Sort the DataFrame by 'EventDate' in ascending order
df_all_sorted = df_all.sort_values(by='EventDate', ascending=True)
df_all_sorted.head()
```

```python
df_all_sorted.County.unique()
```

```python
osceola_data=df_all_sorted[df_all_sorted['County'] == 'Osceola']
osceola_data.head()
```

```python
st_lucie_data=df_all_sorted[df_all_sorted['County'] == 'St. Lucie']
st_lucie_data.head()
```

```python
osceola_data_count=osceola_data.groupby(['EventDate']).count()
osceola_data_count.head()
```

```python
st_lucie_data_count=st_lucie_data.groupby(['EventDate']).count()
st_lucie_data_count.head()
```

```python
osceola_data_count['Date'] =osceola_data_count.index.values
osceola_data_count['Daily Cases'] = osceola_data_count['County']
osceola_data_count['Moving Average'] = osceola_data_count['County'].rolling(5, min_periods=0).mean()
osceola_data_count.head()
```

```python
st_lucie_data_count['Date'] =st_lucie_data_count.index.values
st_lucie_data_count['Daily Cases'] = st_lucie_data_count['County']
st_lucie_data_count['Moving Average'] = st_lucie_data_count['County'].rolling(5, min_periods=0).mean()
st_lucie_data_count.head()
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(15, 10))

# Plotting for Osceola County (first row)
axes[0, 0].plot(osceola_data_count['Date'], osceola_data_count['Daily Cases'], label='Daily Cases', color='blue')
axes[0, 0].set_title('Osceola County - Daily Cases')
axes[0, 0].set_xlabel('Date')
axes[0, 0].set_ylabel('Cases')

axes[0, 1].plot(osceola_data_count['Date'], osceola_data_count['Moving Average'], label='Moving Average', color='red')
axes[0, 1].set_title('Osceola County - Moving Average')
axes[0, 1].set_xlabel('Date')
axes[0, 1].set_ylabel('Cases')

# Plotting for St. Lucie County (second row)
axes[1, 0].plot(st_lucie_data_count['Date'], st_lucie_data_count['Daily Cases'], label='Daily Cases', color='blue')
axes[1, 0].set_title('St. Lucie County - Daily Cases')
axes[1, 0].set_xlabel('Date')
axes[1, 0].set_ylabel('Cases')

axes[1, 1].plot(st_lucie_data_count['Date'], st_lucie_data_count['Moving Average'], label='Moving Average', color='red')
axes[1, 1].set_title('St. Lucie County - Moving Average')
axes[1, 1].set_xlabel('Date')
axes[1, 1].set_ylabel('Cases')

#display the plot

plt.show()
```
